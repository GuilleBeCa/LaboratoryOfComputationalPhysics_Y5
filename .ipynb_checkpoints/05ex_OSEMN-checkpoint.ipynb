2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
random_num=np.random.random(25)
random_matrix=random_num.reshape(5,5)
```

```python
with open('data.txt','w') as f:
    for line in random_matrix:
        for item in line:
            f.write((str(item)+' '))
        
```

3\. Load the saved txt file of point 2 and convert it to a csv file (by hand)

```python
lista=[]
with open('data.txt', 'r') as f:
    for line in f:
        nums=line.split(' ')[:-1]
        for i in nums:
            lista.append(float(i))
random_matrix_2=np.array(lista).reshape(5,5)
```

```python
random_matrix==random_matrix_2
```

```python
import csv
```

```python
with open('data.csv', 'w') as f:
        writer=csv.writer(f)
        writer.writerow(random_matrix_2)
        
```

4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
chr(int('110111   ',2))
```

```python
list_credit=[]
with open('credit_card.dat', 'r') as f:
    word_counter=0
    word_len=6
    
    for line in f:
        a=0
        credit_line=[]
        for i in range(word_len, len(line), word_len):
            new_char=chr(int(line[a:i],2))
            a=i
            credit_line.append(new_char)
        print(credit_line)
        list_credit.append(credit_line)
            
```
